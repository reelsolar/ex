median filtering and histogram

```python
im = skimage.io.imread('//rsi-el/Users/Public/EL/2016-12-05/63J2B.jpg', flatten=True)

# Make the structuring element
selem = skimage.morphology.square(3)

# Perform the median filter
im_filt = skimage.filters.median(im, selem)

hist, bins = skimage.exposure.histogram(im)


# Show filtered image
with sns.axes_style('white'):
    fig, ax = plt.subplots(1, 2, figsize=(14,5))
    ax[0].imshow(im_filt[240:1980,760:2540], cmap=plt.cm.RdBu_r)
    ax[1].fill_between(bins, hist, alpha=0.5)
    ax[1].set_title('Histogram of grey values')
    plt.yscale('log')
    
```

background subtraction

```python
# Convert the uneven image to floating point
im_float = im_filt / im_filt.max()

# Strongly blur the image with a Gaussian filter with stdev of 20 pixels
im_blur = skimage.filters.gaussian_filter(im_float, 20)

# Subtract the blurred image from the original
im_new = im_float - im_blur

# Display images in false color
with sns.axes_style('white'):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    ax[0].imshow(im_float, cmap=plt.cm.RdBu_r)
    ax[1].imshow(im_new, cmap=plt.cm.RdBu_r)
    ax[0].set_title('original')
    ax[1].set_title('background subtracted')
```

local maxima

```python
from scipy import ndimage as ndi
from skimage.feature import peak_local_max
from skimage import data, img_as_float

im_filt = skimage.filters.median(im, selem)

# image_max is the dilation of im with a 20*20 structuring element
# It is used within peak_local_max function
image_max = ndi.maximum_filter(im_filt[240:1980,760:2540], size=20, mode='constant')

# Comparison between image_max and im to find the coordinates of local maxima
coordinates = peak_local_max(im_filt[240:1980,760:2540], min_distance=20)

# display results
fig, axes = plt.subplots(1, 3, figsize=(20, 9), sharex=True, sharey=True,
                         subplot_kw={'adjustable': 'box-forced'})
ax = axes.ravel()
ax[0].imshow(im_filt[240:1980,760:2540], cmap=plt.cm.gray)
ax[0].axis('off')
ax[0].set_title('Original')

ax[1].imshow(image_max, cmap=plt.cm.gray)
ax[1].axis('off')
ax[1].set_title('Maximum filter')

ax[2].imshow(im_filt[240:1980,760:2540], cmap=plt.cm.gray)
ax[2].autoscale(False)
ax[2].plot(coordinates[:, 1], coordinates[:, 0], 'r.')
ax[2].axis('off')
ax[2].set_title('Peak local max')

fig.tight_layout()

plt.show()
```
